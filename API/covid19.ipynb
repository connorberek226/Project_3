```python
import json
import requests
from pprint import pprint
import matplotlib.pyplot as plt 
import pandas as pd
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect
import sqlite3
```

```python
#save config information
url = "https://data.cdc.gov/resource/9mfq-cb36.json"
```

```python
covid19 = requests.get(url)
```

```python
covid19_json = covid19.json()
```

```python
#print (covid19_json)
```

```python
#pprint (covid19_json)
```

```python
covid_dataframe = pd.DataFrame (covid19_json)
```

```python
from datetime import datetime
covid_dataframe["submission_date"] = pd.to_datetime(covid_dataframe["submission_date"], format='%Y-%m-%d %H:%M:%S.%f')
covid_dataframe
```

```python
engine = create_engine('sqlite:///covid19.db', echo=True)
```

```python
sqliteConnection = engine.connect()
```

```python
df = covid_dataframe
```

```python
df.to_sql('covid', sqliteConnection,if_exists='replace')
```

```python
engine.execute("SELECT * FROM covid").fetchall()
```

```python
sqliteConnection.close()
```
